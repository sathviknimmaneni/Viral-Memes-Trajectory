```python
import igraph as ig
import random

G = ig.Graph()

G.add_vertices(36)

G.add_edges([(0,1),(1,3),(2,3),(3,5),(4,5),(5,6),(6,8),(7,8),(8,13),(9,13),(10,13),(11,13),(12,13),(13,14),(14,15),(15,16),(14,17),(17,19),(19,18),(19,22),(19,26),(22,20),(22,21),(22,23),(23,24),(24,25),(23,25),(22,25),(25,27),(26,27),(26,28),(27,29),(26,29),(28,29),(27,28),(28,30),(29,30),(30,31),(31,32),(32,33),(31,33),(33,35),(33,34),(7,35)])
```

```python
K_shell = G.shell_index()
```

```python
[x for x in G.vs.indices if K_shell[x] == max(K_shell)]
```

```python
Deg = G.degree()
```

```python
current = 0
target = 26
```

```python
path = []
path.append(current)
```

### Find neighbors and given algorithm implementation

```python
while (current != target):
    next_nodes = G.neighbors(current)
    req = [x for x in next_nodes if x not in path]
    max_shell = max([K_shell[x] for x in req])
    current_shell = K_shell[current]
    if (max_shell <= current_shell):
        max_degree = max([Deg[x] for x in req])
        rand = [x for x in req if Deg[x] == max_degree]
        current = random.choice(rand)
        path.append(current)
    else:
        rand = [x for x in req if K_shell[x] == max_shell]
        current = random.choice(rand)
        path.append(current)
```

# Shell based hill climbing

```python
import igraph as ig
G = ig.Graph()

G.add_vertices(36)
G.add_edges([(0,1),(1,3),(2,3),(3,5),(4,5),(5,6),(6,8),(7,8),(8,13),(9,13),(10,13),(11,13),(12,13),(13,14),(14,17),(17,19),(19,18),(19,22),(22,20),(22,21),(22,23),(23,24),(24,25),(23,25),(22,25),(25,27),(26,27),(26,28),(27,29),(26,29),(28,29),(27,28),(28,30),(29,30),(30,31),(31,32),(32,33),(31,33),(33,35),(33,34),(7,35)])
K = G.shell_index() #k shell decomposition
D = G.degree()  #degree of nodes
current = 0
path = []
path.append(current)
while K[current]<max(K): 
    neighbor = G.neighbors(current)
    ns = [x for x in neighbor if x not in path]
    y=[]
    for x in range(0,len(ns)): #finds nighbor's degree
        y.append(D[ns[x]])
    if y==[]:   #if node has no unvisted neighbors it breaks
        break
    i= y.index(max(y))  #finds the node with highest degree
    current= ns[i]     
    path.append(current)
print(path)

for x in path:   #plotting the final path
    G.vs[x]['color'] = 'blue'
ig.plot(G, vertex_label=list(range(35))).show()
```

```python
import igraph as ig
import random
import pandas as pd

df = pd.read_csv('facebook_combined.txt',sep=' ',names = ['a','b']) 

tuples = [tuple(x) for x in df.values]
G = ig.Graph.TupleList(tuples)

K_shell = G.shell_index()
Deg = G.degree()

current = 11
target = 352

path = []
path.append(current)

while (current != target):
    next_nodes = G.neighbors(current)
    req = [x for x in next_nodes if x not in path]
    max_shell = max([K_shell[x] for x in req])
    current_shell = K_shell[current]
    if (max_shell <= current_shell):
        max_degree = max([Deg[x] for x in req])
        rand = [x for x in req if Deg[x] == max_degree]
        current = rand.argmax()
        path.append(current)
    else:
        rand = [x for x in req if K_shell[x] == max_shell]
        current = random.choice(rand)
        path.append(current)
```
